## Creating a column that groups by the age 

```python
import pandas as pd
import numpy as np

# Assuming your DataFrame is named users_df

# 1. Define the bin edges (the cut-off points)
# The bins go from the lowest observed age (or 0) up to 100.
bins = [15, 20, 30, 40, 50, 60, 70, 80, 90, 100]

# 2. Define the labels for the resulting age groups
labels = [
    '15-20', 
    '21-30', 
    '31-40', 
    '41-50', 
    '51-60', 
    '61-70', 
    '71-80', 
    '81-90', 
    '91-100'
]

# 3. Use pd.cut() to create the new column
# 'right=True' means the bins include the rightmost edge (e.g., up to 30)
users['age_group'] = pd.cut(
    users['age'], 
    bins=bins, 
    labels=labels, 
    right=True,
    include_lowest=True # Includes the lowest edge (15)
)

# 4. Verification: Check the breakdown of the new column
print("### Verification of New Age Group Column ###")
print(users['age_group'].value_counts().sort_index())
print("\nSample Data:")
print(users[['age', 'age_group']].head(10))
```

## Question 15
Total Churn Proxy Rate and one-month vs two-month segment breakdown

```python
eng_flags = engagement[["user_id", "left_for_one_month", "left_for_two_month_plus"]].copy()

# User considered churned if either flag is 1
eng_flags["churn_proxy"] = (
    (eng_flags["left_for_one_month"] == 1) |
    (eng_flags["left_for_two_month_plus"] == 1)
).astype(int)

# Aggregate to user level in case there are multiple engagement rows per user
user_churn = eng_flags.groupby("user_id", as_index=False)["churn_proxy"].max()

# Get all distinct users from users table
all_users = users[["user_id"]].drop_duplicates()

# Left join churn info so every user has a churn flag
user_churn_full = all_users.merge(user_churn, on="user_id", how="left")
user_churn_full["churn_proxy"] = user_churn_full["churn_proxy"].fillna(0).astype(int)

# Total churn proxy rate
total_users = len(user_churn_full)
churned_users = int(user_churn_full["churn_proxy"].sum())
churn_proxy_rate = churned_users / total_users if total_users > 0 else float("")

print("Total users:", total_users)
print("Churned users (proxy):", churned_users)
print("Total Churn Proxy Rate:", churn_proxy_rate)

# Segment breakdown: one-month-only vs any two-month-plus vs no flag

user_flags_agg = eng_flags.groupby("user_id", as_index=False)[
    ["left_for_one_month", "left_for_two_month_plus"]
].max()

one_month_only = ((user_flags_agg["left_for_one_month"] == 1) &
                  (user_flags_agg["left_for_two_month_plus"] == 0)).sum()

two_month_plus_any = (user_flags_agg["left_for_two_month_plus"] == 1).sum()

no_flag = ((user_flags_agg["left_for_one_month"] == 0) &
           (user_flags_agg["left_for_two_month_plus"] == 0)).sum()

print("Segment counts")
print("One-month only:", int(one_month_only))
print("Two-month-plus:", int(two_month_plus_any))
print("No churn flag:", int(no_flag))
```

## Question 17
How many users are retained vs. churned, and how do we define a retention label from the churn flags?

```python
import pandas as pd

# Assume engagement, transactions, users are already loaded dataframes.
# If not, you would load them here with pd.read_csv or similar.

# 1. Build a user-level churn / retention label
eng_flags = engagement[["user_id", "left_for_one_month", "left_for_two_month_plus"]].copy()

# Ensure numeric 0/1 flags
for col in ["left_for_one_month", "left_for_two_month_plus"]:
    eng_flags[col] = pd.to_numeric(eng_flags[col], errors="coerce").fillna(0)

# Aggregate at user level (in case multiple engagement rows per user)
eng_flags_user = eng_flags.groupby("user_id", as_index=False)[
    ["left_for_one_month", "left_for_two_month_plus"]
].max()

# Define a churn proxy: churn if 1-month OR 2+ months flag is 1
eng_flags_user["churned"] = (
    (eng_flags_user["left_for_one_month"] == 1) |
    (eng_flags_user["left_for_two_month_plus"] == 1)
).astype(int)

# Define retention: retained = 1 - churned
eng_flags_user["retained"] = 1 - eng_flags_user["churned"]

# Basic counts
retention_counts = eng_flags_user["retained"].value_counts(dropna=False)

print("Retention label counts (1 = retained, 0 = churned)")
print(retention_counts)
```

## Question 18
How is retention related to basic app usage features (app_downloaded, web_user, ios_user, android_user, app_web_user)?

```python
# 2. Merge user-level retention with engagement app flags
app_flags_cols = ["user_id", "app_downloaded", "web_user", "app_web_user", "ios_user", "android_user"]

missing_cols = [c for c in app_flags_cols if c not in engagement.columns]
if missing_cols:
    print("Missing app feature columns in engagement:", missing_cols)

eng_app_flags = engagement[[c for c in app_flags_cols if c in engagement.columns]].copy()

# Aggregate app flags to user level (max works for 0/1 flags)
eng_app_flags_user = eng_app_flags.groupby("user_id", as_index=False).max()

# Merge with retention labels
user_retention_app = eng_flags_user.merge(eng_app_flags_user, on="user_id", how="left")

# For each app feature, compute retention rate among users with that feature = 1
app_feature_cols = [c for c in app_flags_cols if c != "user_id" and c in user_retention_app.columns]

retention_by_feature = {}

for col in app_feature_cols:
    has_feature = user_retention_app[user_retention_app[col] == 1]
    if len(has_feature) > 0:
        retention_rate = has_feature["retained"].mean()
        retention_by_feature[col] = retention_rate

print("Retention rate among users with given app feature = 1")
print(retention_by_feature)
```

## Question 19
What is the correlation between app usage features and the retention label?

```python
corr_app = {}

for col in app_feature_cols:
    # Pearson correlation between retained (0/1) and feature (0/1)
    corr_val = user_retention_app[["retained", col]].corr().loc["retained", col]
    corr_app[col] = corr_val

print("Correlation between app features and retention (1 = retained):")
print(corr_app)
```

```python
# This cell trains a Logistic Regression model (for churn probability)
# and a Random Forest model (also for churn probability), and evaluates both.

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score, classification_report
import numpy as np

# ----- 1) Define features X and target y -----
exclude_cols = ["user_id", "churn"] + churn_flag_cols
feature_cols = [c for c in model_df.columns if c not in exclude_cols]

X = model_df[feature_cols].fillna(0)
y = model_df["churn"].astype(int)

# ----- 2) Train/test split -----
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# ----- 3) Logistic Regression (with scaling) -----
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

log_reg = LogisticRegression(
    max_iter=1000,
    solver="lbfgs",
    n_jobs=-1
)

log_reg.fit(X_train_scaled, y_train)

y_proba_lr = log_reg.predict_proba(X_test_scaled)[:, 1]
y_pred_lr = (y_proba_lr >= 0.5).astype(int)

auc_lr = roc_auc_score(y_test, y_proba_lr)
report_lr = classification_report(y_test, y_pred_lr)

print("=== Logistic Regression Performance ===")
print("AUC:")
print(auc_lr)
print("Classification report:")
print(report_lr)

# ----- 4) Random Forest for prediction and comparison -----
rf = RandomForestClassifier(
    n_estimators=300,
    max_depth=None,
    min_samples_split=10,
    min_samples_leaf=5,
    random_state=42,
    n_jobs=-1
)

rf.fit(X_train, y_train)

y_proba_rf = rf.predict_proba(X_test)[:, 1]
y_pred_rf = rf.predict(X_test)

auc_rf = roc_auc_score(y_test, y_proba_rf)
report_rf = classification_report(y_test, y_pred_rf)

print("=== Random Forest Performance ===")
print("AUC:")
print(auc_rf)
print("Classification report:")
print(report_rf)
```
